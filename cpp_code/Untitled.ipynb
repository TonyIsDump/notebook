```python
import numpy as np
import matplotlib.pyplot as plt
from math import*

def f(t, x):
    x1, x2 = x
    dx1 = x2
    dx2 = -9.81*sin(x1)
    return np.array([dx1, dx2])

def rk4_step(t, x, h, f):
    k1 = f(t, x)
    k2 = f(t + 0.5*h, x + 0.5*h*k1)
    k3 = f(t + 0.5*h, x + 0.5*h*k2)
    k4 = f(t + h, x + h*k3)
    return x + (h/6.0)*(k1 + 2*k2 + 2*k3 + k4)

# Parameters
t0 = 0.0
t_end = 5.0
h = 0.0001  # step size

# Time grid
N = int((t_end - t0)/h)
t = np.linspace(t0, t_end, N+1)

# Initial conditions: y(0) = 0, y'(0) = 5
x = np.zeros((N+1, 2))
x[0, 0] = 0.0  # y
x[0, 1] = 10.0  # y'

# Integrate with RK4
for n in range(N):
    x[n+1] = rk4_step(t[n], x[n], h, f)

y = x[:, 0]

# Plot
plt.plot(t, y, linewidth=2)
plt.xlabel('t')
plt.ylabel('y')
plt.title("Solution of y'' + 6y' + 5y = 10 with RK4")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
